```python
import discretisedfield as df
import discretisedfield.tests as dt
import matplotlib.pyplot as plt
from discretisedfield.tools import topological_charge
```

```python
top_charges = list()
for i in range(0, 1000):
    f = df.Field.fromfile(f'./Ovf-files-for-tmig/m{i:06d}.ovf')
    t = topological_charge(f.project('z'))
    top_charges.append(t)
    print(f'frame number:{i}, topological_charge{t}')
```
