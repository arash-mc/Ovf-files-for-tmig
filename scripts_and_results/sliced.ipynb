```python
top_charges = list()
for i in range(0, 1000):
    f = df.Field.fromfile(f'./Ovf-files-for-tmig/m{i:06d}.ovf')
    t = topological_charge(f.plane('z'))
    top_charges.append(t)
    print(f'frame number:{i}, topological_charge{t}')
```

```python
top_single_plane = {i:top_charges[i] for i in range(1000)}
with open('top_single_plane.txt', 'w') as file:
    file.write(str(top_single_plane))
```
